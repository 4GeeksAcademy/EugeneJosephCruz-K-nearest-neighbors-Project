```python
import pandas as pd  # Import the pandas library and give it the alias 'pd' for easier access.

# Load a dataset of movies from a specified URL into a pandas DataFrame called 'movies'.
# The dataset is in CSV (Comma-Separated Values) format.
movies = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_movies.csv")

# Load a dataset of movie credits from another specified URL into a pandas DataFrame called 'credits'.
# This dataset also is in CSV format and includes information like cast and crew for the movies.
credits = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_credits.csv")

# Display the first 5 rows of the 'movies' DataFrame to provide a quick overview of the data it contains.
# This is useful for verifying the structure of the DataFrame, the type of data it holds, and if it was loaded correctly.
movies.head()
```

```python
# Display the first 5 rows of the 'credits' DataFrame to provide a quick overview of the data it contains.
# This is useful for checking the structure of the DataFrame, the type of data it holds, and ensuring it was loaded correctly.
credits.head()
```

```python
import sqlite3  # Import the sqlite3 library to interact with SQLite databases.

# Establish a connection to a SQLite database. The database is located in the 'data' directory,
# and is named 'movies_database.db'. If the database does not exist, it will be created.
conn = sqlite3.connect("../data/movies_database.db")

# Convert the 'movies' DataFrame into a SQL table within the connected SQLite database.
# The table will be named 'movies_table'. If this table already exists, it will be replaced.
# 'index = False' indicates that the DataFrame's index will not be written to the SQL table.
movies.to_sql("movies_table", conn, if_exists = "replace", index = False)

# Similarly, convert the 'credits' DataFrame into a SQL table named 'credits_table' within the same database.
# The table is replaced if it already exists, and the DataFrame's index is not included in the SQL table.
credits.to_sql("credits_table", conn, if_exists = "replace", index = False)
```

```python
# Define a SQL query to merge two tables: 'movies_table' and 'credits_table'.
# The merging is done through an INNER JOIN based on the 'title' column, meaning only rows with matching titles in both tables will be included.
query = """
    SELECT *
    FROM movies_table
    INNER JOIN credits_table
    ON movies_table.title = credits_table.title;
"""

# Execute the SQL query using the established database connection 'conn'.
# The result of this query is stored in a new DataFrame called 'total_data'.
# This DataFrame now contains combined information from both the 'movies' and 'credits' DataFrames based on movie titles.
total_data = pd.read_sql_query(query, conn)

# Close the connection to the SQLite database now that the necessary data has been retrieved.
# This is important to free up database resources and avoid potential database locks.
conn.close()

# Remove duplicate columns from the 'total_data' DataFrame.
# 'total_data.columns.duplicated()' returns a boolean array where True indicates a duplicate column.
# The '~' operator inverts the boolean values, so True becomes False and vice versa.
# 'total_data.loc[:, ~total_data.columns.duplicated()]' selects all rows and only columns that are not duplicated.
total_data = total_data.loc[:, ~total_data.columns.duplicated()]

# Display the first 5 rows of the 'total_data' DataFrame to provide an overview of the merged data,
# ensuring that the merge was successful and the dataset is as expected.
total_data.head()
```

```python
import json  # Import the json library for parsing JSON data.

# Define a function that attempts to load a JSON string safely.
# If the JSON loading fails due to a TypeError or a JSONDecodeError, it returns a default value instead.
def load_json_safe(json_str, default_value = None):
    try:
        return json.loads(json_str)  # Try to parse the JSON string.
    except (TypeError, json.JSONDecodeError):  # If parsing fails, catch the exception.
        return default_value  # Return the default value in case of an exception.

# Transform the 'genres' column: parse the JSON data and replace it with a list of genre names.
# If the original data is not NaN, it's parsed and the names are extracted; otherwise, None is assigned.
total_data["genres"] = total_data["genres"].apply(lambda x: [item["name"] for item in json.loads(x)] if pd.notna(x) else None)

# Similar transformation is applied to the 'keywords' column, extracting a list of keyword names from the JSON data.
total_data["keywords"] = total_data["keywords"].apply(lambda x: [item["name"] for item in json.loads(x)] if pd.notna(x) else None)

# For the 'cast' column, parse the JSON data, but only keep the names of the first three cast members.
# This is done by slicing the list to the first 3 elements after parsing and extracting names.
total_data["cast"] = total_data["cast"].apply(lambda x: [item["name"] for item in json.loads(x)][:3] if pd.notna(x) else None)

# Transform the 'crew' column to only include the names of the directors.
# It uses the 'load_json_safe' function to parse the JSON and then filters for crew members with the job 'Director'.
total_data["crew"] = total_data["crew"].apply(lambda x: " ".join([crew_member['name'] for crew_member in load_json_safe(x) if crew_member['job'] == 'Director']))

# The 'overview' column is transformed into a list containing the original overview string.
# This could be for consistency in data structure or further processing needs.
total_data["overview"] = total_data["overview"].apply(lambda x: [x])

# Display the first 5 rows of the transformed 'total_data' DataFrame to verify the transformations.
total_data.head()
```

```python
# Ensure each element in the 'overview' list is a string, even if it was originally of a different type.
total_data["overview"] = total_data["overview"].apply(lambda x: [str(x)])

# Convert each genre in the 'genres' list to a string, to ensure consistency and avoid potential type-related errors.
total_data["genres"] = total_data["genres"].apply(lambda x: [str(genre) for genre in x])

# Convert each keyword in the 'keywords' list to a string, for consistency and to handle any non-string types.
total_data["keywords"] = total_data["keywords"].apply(lambda x: [str(keyword) for keyword in x])

# Ensure each actor name in the 'cast' list is converted to a string, to standardize the data format.
total_data["cast"] = total_data["cast"].apply(lambda x: [str(actor) for actor in x])

# Convert each crew member's name in the 'crew' list to a string, to unify the data type across the column.
total_data["crew"] = total_data["crew"].apply(lambda x: [str(crew_member) for crew_member in x])

# Concatenate the lists from 'overview', 'genres', 'keywords', 'cast', and 'crew' columns into a single list in a new column 'tags'.
# Each element of these lists is then joined into a single string, with spaces between each element, effectively creating a 'bag of words'.
total_data["tags"] = total_data["overview"] + total_data["genres"] + total_data["keywords"] + total_data["cast"] + total_data["crew"]
total_data["tags"] = total_data["tags"].apply(lambda x: ",".join(x).replace(",", " "))

# Drop the original columns ('genres', 'keywords', 'cast', 'crew', 'overview') from 'total_data',
# as their information has been consolidated into the 'tags' column.
total_data.drop(columns = ["genres", "keywords", "cast", "crew", "overview"], inplace = True)

# Display the 'tags' content for the first row of the 'total_data' DataFrame.
# This shows the combined information of the dropped columns as a single string, which might be used for text analysis or as metadata.
total_data.iloc[0].tags
```

```python
# Export the 'total_data' DataFrame to a CSV file, effectively saving the processed and transformed data.
# The file is saved to the 'processed' subdirectory within 'data', and is named 'clean_data.csv'.
# 'index = False' ensures that the DataFrame index is not included in the CSV file, only the data.
total_data.to_csv("../data/processed/clean_data.csv", index = False)

# Re-establish a connection to the SQLite database located in the 'data' directory, named 'movies_database.db'.
# This connection is used to save the transformed data back into the database.
conn = sqlite3.connect("../data/movies_database.db")

# Convert the 'movies' DataFrame (which may still contain the original unprocessed data) into a SQL table within the database.
# The table is named 'clean_movies_data'. If this table already exists, it is replaced with the current version of the 'movies' DataFrame.
# 'index = False' specifies that the DataFrame's index should not be included as a column in the SQL table.
movies.to_sql("clean_movies_data", conn, if_exists = "replace", index = False)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer  # Import TfidfVectorizer for text vectorization.
from sklearn.neighbors import NearestNeighbors  # Import NearestNeighbors for the KNN model.

# Initialize a TfidfVectorizer object to convert the text data into a matrix of TF-IDF features.
vectorizer = TfidfVectorizer()
# Fit the vectorizer to the 'tags' column of 'total_data' and transform the text into a TF-IDF matrix.
tfidf_matrix = vectorizer.fit_transform(total_data["tags"])

# Initialize a NearestNeighbors model specifying the number of neighbors, the algorithm to use, and the metric.
# 'n_neighbors = 6' means the model will find the 6 nearest neighbors.
# 'algorithm = "brute"' specifies to use the brute-force search.
# 'metric = "cosine"' sets the cosine similarity as the distance metric to measure similarity between vectors.
model = NearestNeighbors(n_neighbors = 6, algorithm = "brute", metric = "cosine")
# Fit the KNN model to the TF-IDF matrix, effectively preparing it to find nearest neighbors in this feature space.
model.fit(tfidf_matrix)

# Define a function to get movie recommendations based on the title of a movie.
def get_movie_recommendations(movie_title):
    # Find the index of the movie in 'total_data' that matches the given title.
    movie_index = total_data[total_data["title"] == movie_title].index[0]
    # Use the KNN model to find the distances and indices of the 6 nearest neighbors to the given movie.
    distances, indices = model.kneighbors(tfidf_matrix[movie_index])
    # Construct a list of similar movies, excluding the first one (which is the input movie itself).
    similar_movies = [(total_data["title"][i], distances[0][j]) for j, i in enumerate(indices[0])]
    return similar_movies[1:]  # Return the list of recommendations, excluding the input movie.

# Specify a movie title to get recommendations for.
input_movie = "How to Train Your Dragon"
# Get the list of recommended movies for the input title using the defined function.
recommendations = get_movie_recommendations(input_movie)

# Print out the recommendations.
print("Film recommendations for '{}'".format(input_movie))
for movie, distance in recommendations:
    # Each recommendation includes the title of the movie and its distance (a measure of similarity) from the input movie.
    print("- Film: {}".format(movie))
```

```python
from pickle import dump  # Import the 'dump' function from the 'pickle' module for serializing objects.

# Serialize the trained KNN model ('model') and save it to a file using 'dump'.
# The file is saved in the 'models' directory with the name 'knn_neighbors-6_algorithm-brute_metric-cosine.sav'.
# 'wb' mode is used to write in binary format, which is required for pickling.
dump(model, open("../models/knn_neighbors-6_algorithm-brute_metric-cosine.sav", "wb"))
```
